----------------------------- Combine e Modify Driver -----------------------------------------------------------------------------------------------------------------------------------

```python
driver_standings
```

```python
driver_standings = driver_standings.drop(['driverStandingsId','positionText'], axis=1)
driver_standings
```

```python
drivers = drivers.drop(['driverRef','url'],axis=1)
drivers
```

```python
drivers_table = pd.merge(drivers,driver_standings,how='inner',on='driverId')
drivers_table
```

```python
drivers_brazil = drivers_table[drivers_table['nationality']=='Brazilian']
```

```python
drivers_brazil
```

```python
group_brazil = drivers_brazil.groupby('surname',as_index=False)
group_brazil = group_brazil.sum()
group_brazil
```

```python
group_brazil = group_brazil.sort_values(by=['wins'],ascending=False, ignore_index=True)
group_brazil
```

```python
group_brazil = group_brazil.drop(group_brazil.index[6:])
group_brazil = group_brazil.drop(['driverId','raceId','position'],axis=1)
group_brazil
```

```python
plt.figure(figsize = (20, 7))
sns.barplot(x = group_brazil["surname"], y = group_brazil["wins"], palette = "autumn");
```

```python
plt.figure(figsize = (20, 6))
plt.ticklabel_format(style = 'plain')
sns.barplot(x = group_brazil["points"], y = group_brazil["surname"], palette = "nipy_spectral");
```

----------------------------- Combine e Modify PitStop -----------------------------------------------------------------------------------------------------------------------------------

```python
pit_stops
```

```python
pit_stops_nationality = pd.merge(pit_stops,drivers_brazil,how='left',on='driverId')
pit_stops_nationality= pit_stops_nationality.dropna(how='any',axis=0)
pit_stops_nationality
```

```python
pit_stops_brazil = pit_stops_nationality[pit_stops_nationality['nationality']=='Brazilian']
pit_stops_brazil
```

```python
pit_stops_drivers = pit_stops_brazil.groupby('surname',as_index=False)
pit_stops_drivers = pit_stops_drivers.mean()
pit_stops_drivers
```

```python
plt.figure(figsize = (20, 6))
plt.ticklabel_format(style = 'plain')
plt.title('Mean Brazilians Pilot PitStop')
plt.xlabel('Mean Miliseconds')
sns.barplot(x = pit_stops_drivers["milliseconds"], y = pit_stops_drivers["surname"], palette = "rocket")
```
